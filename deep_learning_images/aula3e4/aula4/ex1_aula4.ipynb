# EX1 - Aula 4 - Visualização dos Filtros da CNN

```python
import keras
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from keras import models

model = models.load_model("cnn1.h5")
```

```python
(filters, biases) = model.get_layer("conv2d_1").get_weights()
```

```python
filters.shape
```

```python
plt.figure(figsize=(20,10))

for i in range(0, filters.shape[3]):
    
    plt.subplot(4,5,i+1)    # the number of images in the grid is 5*5 (25)
    
    img = filters[:,:,0,i]
    
    plt.imshow(img)
    
plt.colorbar()
    
plt.show()
```
