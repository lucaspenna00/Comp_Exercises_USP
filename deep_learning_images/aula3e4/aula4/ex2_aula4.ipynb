# 1. Função que dada a imagem, modelo e camada plota as ativações

```python
def get_predictions_intermediate_layer(img, modelo, layer_name):
    
    from keras import models, Model
    import matplotlib.pyplot as plt
    import numpy as np
    
    model = models.load_model(modelo)
    
    img = img / 255.0
    
    img = np.expand_dims(img,axis=0)
    
    img = np.expand_dims(img,axis=3)
    
    intermediate_layer_model = Model(inputs=model.input, outputs=model.get_layer(layer_name).output)

    y = intermediate_layer_model.predict(img)
    
    plt.figure(figsize=(20,10))

    for i in range(0, y.shape[3]):

        plt.subplot(4,5,i+1)    # the number of images in the grid is 5*5 (25)

        img = y[0,:,:,i]

        plt.imshow(img)

    plt.colorbar()

    plt.show()
```
