## Ex1 Utilizar Xception para classificar imagens

```python
import tensorflow.keras as keras
from keras.preprocessing import image
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

## Xception Architecture

<img src= Xception.png>

```python
def xception_predict(img_path):
    
    from keras.applications.xception import Xception
    
    model = Xception()
    
    target_size = (299, 299)
    
    img = image.load_img(img_path, target_size=target_size)
    
    plt.imshow(img)
    
    from keras.applications.xception import preprocess_input

    x = image.img_to_array(img)
    
    x = np.expand_dims(x, axis=0)
        
    x = preprocess_input(x)
    
    preds = model.predict(x)
    
    from keras.applications.xception import decode_predictions

    p = decode_predictions(preds, top=3)[0]
    
    for predicao in p:
        
        print("%8.2f%% %s"%(100*predicao[2],predicao[1]))
```

```python
xception_predict('terrier.jpg')
```

```python
xception_predict('onibus.jpg')
```

```python
xception_predict('carro.jpg')
```

```python
xception_predict('predio.jpg')
```
